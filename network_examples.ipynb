```python
G_weighted = nx.Graph()
G_weighted.add_edge('A','B', weight=25)
G_weighted.add_edge('A','D', weight=8)
G_weighted.add_edge('A','E', weight=11)
G_weighted.add_edge('A','C', weight=1)
G_weighted.add_edge('B','D', weight=4)
G_weighted.add_edge('B','E',weight=7)
G_weighted.add_edge('B','C', weight=1)
G_weighted.add_edge('C','D',weight=1)

pos = nx.circular_layout(G_weighted)

edges = G_weighted.edges()
#colors = [G_weighted[u][v]['color'] for u,v in edges]
weights = [G_weighted[u][v]['weight']/5 for u,v in edges]

#nx.draw(G_weighted, pos, edges=edges, edge_color=colors, width=weights)
nx.draw(G_weighted, pos, edges=edges, width=weights, node_color='c', with_labels=True)
#%config InlineBackend.figure_format = 'svg'
```
